# **Visualización de las reseñas**

```python

import re

MODEL_OPTIONS = ['gpt-3.5-turbo-16k', 'gpt-3.5-turbo', 'gpt-4']

def ask_gpt(model_choice: str, imdb_reviews, review_index: int) -> str:
    try:
        selected_review = imdb_reviews[review_index]['text']
        prompt = f"Reseña de la película Avatar: {selected_review}"
        response = openai.ChatCompletion.create(
            model=model_choice,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ]
        )
        return response['choices'][0]['message']['content'].strip()
    except Exception as e:
        return f"Ocurrió un error al procesar la solicitud: {e}"

# Solicitar al usuario que elija un modelo
while True:
    model_choice = input(f"Por favor, elija un modelo ({', '.join(MODEL_OPTIONS)}): ")
    if model_choice in MODEL_OPTIONS:
        break
    print("La opción seleccionada no es válida. Intente nuevamente.")

# Filtrar reseñas de la película escrita
movie_title = "Avatar"
pattern = re.compile(re.escape(movie_title), re.IGNORECASE)
imdb_reviews = [review for review in train_dataset if pattern.search(review["text"])]

# Seleccionar una reseña al azar
import random
review_index = random.randint(0, len(imdb_reviews))

# Generar respuesta del modelo
response = ask_gpt(model_choice, imdb_reviews, review_index)
print(response)
```

**Cargar el Modelo Preentrenado**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)  # 3 clases: positivo, negativo, neutral
```

**Pre-procesar los datos**

```python
def preprocess_data(data):
    return tokenizer(data['text'], padding=True, truncation=True)

train_data = twitter_dataset['train'].map(preprocess_data)
test_data = twitter_dataset['test'].map(preprocess_data)
validation_data = twitter_dataset['validation'].map(preprocess_data)
```

**Fine-tuning del modelo**

```python
import os
from transformers import TrainingArguments

# Verificar y crear el directorio de salida si no existe
output_dir = '/content/drive/MyDrive/ProyectoTransformer'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

training_args = TrainingArguments(
    output_dir=output_dir,  # Directorio de salida para guardar los checkpoints y los resultados
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    logging_dir='/content/drive/MyDrive/ProyectoTransformer/logs',  # Ruta absoluta para el directorio de registro
    logging_steps=100,
    evaluation_strategy="epoch",
    save_strategy="epoch",  # Cambiamos la estrategia de guardado a 'epoch'
    report_to="tensorboard",
    load_best_model_at_end=True
)
```

**Proceso de entrenamiento**

```python
from transformers import Trainer

# 1. Crear el Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=validation_data,
    tokenizer=tokenizer,
)

# 2. Entrenar el modelo
trainer.train()

# 3. Evaluar el modelo (opcional)
eval_results = trainer.evaluate()

# 4. Guardar el modelo entrenado
model.save_pretrained('/content/drive/MyDrive/ProyectoTransformer/modelo_entrenado')
```
